```python
import pandas as pd
```

```python
movies = pd.read_csv('movies_metadata.csv')
```

```python
movies.shape[0]
```

```python
class PipeLine():
    def __init__(self, movies):
        self.df = movies
    def drop_rows(self):
        self.df = self.df.drop(columns=['belongs_to_collection', 'homepage', 'budget', 'imdb_id', 'id', 'poster_path', 'production_countries', 'revenue', 'release_date', 'status', 'title', 'video', 'budget'], axis='columns')
    def get_df(self):
        return self.df
    def drop_na(self):
        self.df.drop(self.df.loc[self.df['genres'] == '[]'].index.values[0])
        self.df.dropna(axis=0, inplace=True)
    def drop_not_en(self):
        self.df.drop(self.df.loc[self.df['original_language'] != 'en'].index, inplace=True)
    def reset_index(self):
        self.df = self.df.reset_index(drop=True)
```

```python
pipe = PipeLine(movies)
pipe.drop_rows()
pipe.drop_na()
pipe.drop_not_en()
pipe.reset_index()
df = pipe.get_df()
```
